```python
from opensoundscape.audio import Audio
from opensoundscape.spectrogram import Spectrogram

from pathlib import Path
import pandas as pd
```

```python
audio_a = Path('../samples/sample_a.wav')
audio_b = Path('../samples/sample_b.wav')
audio_c = Path('../samples/sample_c.wav')
audio_d = Path('../samples/sample_d.wav')
```

```python
m = 4
size = 2048*m
n = 16*m

audio_object = Audio.from_file(audio_d, sample_rate=48000)
spectrogram_object = Spectrogram.from_audio(audio_object, fft_size=size, window_samples=size/n, overlap_samples=n)
spectrogram_object.plot()
```

```python

image_shape = (size, size*m) #(height, width) not (width, height)
image_path = Path('../output/saved_image.png')
spectrogram_image = spectrogram_object.to_image(shape=image_shape,invert=True)
spectrogram_image.save(image_path)
```
